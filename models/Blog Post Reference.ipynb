```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
housing_data1 = pd.read_csv('kc_house_data.csv')
housing_data1.head()
```

```python
housing_data1 = housing_data1.dropna()
housing_data1 = housing_data1.drop(columns=['lat', 'long', 'yr_renovated', 'view','id', 'zipcode', 'waterfront', 'sqft_basement'])
```

```python
housing_data1.head()
```

```python
sns.set(style="white")

# Create a covariance matrix
corr = housing_data1.corr()

# Generate a mask the size of our covariance matrix
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11,9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220,10,as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr,mask=mask,cmap=cmap,vmax=1,center=0,square=True, 
            linewidth=.5, cbar_kws={'shrink': .5})

ax.set_title('Multi-Collinearity of Features')
plt.savefig('correlation2.png')
```
